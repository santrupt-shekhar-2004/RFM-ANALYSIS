```python
df.drop_duplicates()
```

```python
from datetime import datetime

# Convert 'Date of Birth' to datetime format
df['DOB'] = pd.to_datetime(df['DOB'], errors='coerce')

# Calculate age based on the current date (you may want to replace it with the specific reference date)
current_date = pd.to_datetime('today')
df['Age'] = (current_date - df['DOB']).astype('<m8[Y]')

# Define age groups
bins = [0, 18, 35, 50, 100]
labels = ['0-18', '19-35', '36-50', '51-100']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Display the modified DataFrame
print(df[['DOB', 'Age', 'Age Group']])
```

```python
#deleting rows in dob with null values
for x in df.index:
    if df.loc[x,"DOB"] == '':
        df.drop(x,inplace=True)
```
